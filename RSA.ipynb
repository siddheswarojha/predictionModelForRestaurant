```python
import pandas as pd 
import numpy as nmp
import matplotlib.pyplot as plt
import joblib
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("restaurants.csv")
df
```

```python
df2 = df.drop(['Location','Cuisine'],axis='columns')
df2
```

```python
min_threshold= df2['Votes'].quantile(0.5)

min_threshold
```

```python
max_threshold= df2['Votes'].quantile(0.999)

max_threshold
```

```python
df3 = df2[(df2.Votes>min_threshold)&(df2.Votes<max_threshold)]
df3
```

```python
newDF = pd.get_dummies(df3.City)
newDF
```

```python
finalDF1 = pd.concat([df3,newDF],axis='columns')
finalDF1
```

```python
X = finalDF1.drop(['Name','City','Agra'],axis='columns')
X
```

```python
y = finalDF1.Name
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.02)
```

```python
predictionModel = linear_model.LogisticRegression(solver='lbfgs', max_iter=3000)
predictionModel.fit(X_train,y_train)
```
